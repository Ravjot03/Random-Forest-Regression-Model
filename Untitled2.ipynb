Pre-processing the data

```python
from sklearn.preprocessing import LabelEncoder
label = LabelEncoder()

label.fit(df.sex)
df.sex = label.transform(df.sex)

label.fit(df.smoker)
df.smoker = label.transform(df.smoker)

label.fit(df.region)
df.region = label.transform(df.region)
```

Splitting the data into independent and dependent features

```python
X = df.drop('charges',axis=1)
y = df['charges']
```

```python
X.head()
```

```python
y.head()
```

Splitting the data into train and test phase (Training phase = 70%, Testing phase = 30%)

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y,random_state = 42, test_size=0.3)
```

```python
X_train.shape
```

Importing the Random Forest Regression Model from Scikit-learn

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
```

Initiating the model

```python
RF = RandomForestRegressor(n_estimators=100)
```

Training the model on train data

Testing the model on the test data

Printing the accuracy of our model

```python
from sklearn.metrics import accuracy_score

RF.fit(X_train, y_train)
RF.predict(X_test)
y_pred  = RF.score(X_test, y_test)
y_pred
```

Printing the feature importances for all the features

```python
feature_scores = RF.feature_importances_
feature_scores.sort()
feature_scores
```

Plotting the feature importance against each feature in form of barplot

```python
plt.figure()
plt.barh(X_train.columns, feature_scores)
plt.show()
```

Implementing Grid-search CV

```python
parameters = {
    'n_estimators': [100, 150, 200, 250, 300],
    'max_depth': [1,2,3,4],
}
regr = RandomForestRegressor(random_state=0)

clf = GridSearchCV(regr, parameters)
clf.fit(X_train, y_train)
```

```python
clf.predict(X_test)
y_pred  = clf.score(X_test, y_test)
y_pred
```

Printing the best parameters from the GridSearchCV that has attained maximum accuracy

```python
print(clf.best_params_)
```
